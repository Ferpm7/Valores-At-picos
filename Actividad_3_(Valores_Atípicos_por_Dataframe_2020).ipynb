```python
#Convertir DataFrame a CSV
data1.to_csv("Datos3-2020.csv")
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

###Aplicar métodos (Desviación Estándar y Rango Intercuartílico) para eliminar Outliers, comprobar con diagramas de caja de cada dataframe

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data1
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data2= data1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
data2.to_csv("Datos4-2020.csv")
```

```python
#Eliminamos columnas innecesarias
data3= data2.drop(['FECHA', 'FOLIO', 'GASTO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'DESCRIPCION', 'TIPO','STATUS', 'POLIZA'], axis=1)
data3
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
data3.to_csv("Datos5-2020.csv")
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
data_clean.to_csv("Datos6-2020.csv")
```

```python
#Comprobamos con diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data_clean.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
data5=data2.copy()
data5= data5.drop(['FECHA', 'FOLIO', 'GASTO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'DESCRIPCION', 'TIPO','STATUS', 'POLIZA'], axis=1)

y=data5

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Convertir DataFrame a CSV
data5.to_csv("Datos8-2020.csv")
```

```python
#Obtenemos datos limpios del Dataframe
data5_iqr= data5[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data5_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data5_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr=data5_iqr.copy()
data4_iqr=data5_iqr.fillna(round(data4_iqr.mean(),1))
data4_iqr
```

```python
#Convertir DataFrame a CSV
data4_iqr.to_csv("Datos9-2020.csv")
```

```python
#Comprobamos con diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data4_iqr.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
Base_limpios = data4_iqr
Base_limpios
```

```python
#Convertir DataFrame a CSV
Base_limpios.to_csv("Datos10-2020.csv")
```
